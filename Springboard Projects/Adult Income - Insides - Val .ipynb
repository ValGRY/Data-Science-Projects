# Loading and Cleaning Data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\vikag\Desktop\ONGOING!!!!\adult.csv\adult.csv')
```

```python
df.shape
```

```python
df.head(7).T
```

```python
df.isin(['?']).sum()
```

```python
df['workclass'] = df['workclass'].replace('?', np.nan)
df['occupation'] = df['occupation'].replace('?', np.nan)
df['native-country'] = df['native-country'].replace('?', np.nan)
```

```python
df.dropna(how='any', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
def Fifty_plus(income):
    if income == '<=50K':
        return 0
    else:
        return 1
df['income_2']= df['income'].apply(Fifty_plus)
df.head().T
```

```python
df.info()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# What Distribution Of Races Do We Have?

```python
plt.figure(figsize=(20,10))
total = len(df['race'])
ax = sns.countplot(x='race', data=df)
for h in ax.patches:
    height = h.get_height()
    ax.text(h.get_x() + h.get_width() / 2., height + 3,'{:1.2f}'.format((height/total)*100) + '%',
            ha="center", size=22)
plt.yticks(size = 17)
plt.xticks(size = 17)
plt.ylabel('Number Of Prople', size = 30)
plt.xlabel('Race', size = 30)
plt.title('Distribution Of Races', size = 30)
plt.tight_layout()
```

Looks like 86.02% are white, and only 13.98% other races.

#  What Age Distribution Do We Have?

```python
plt.figure(figsize=(15,6))
plt.yticks(size = 17)
plt.xticks(size = 17)
plt.xlabel('Age', size = 22)
plt.ylabel('Number Of Prople', size = 22)
plt.title('Age Distribution', size = 22)
plt.grid(axis='y', alpha=0.75)
plt.hist(df['age'], bins=20, color = 'skyblue', rwidth=0.85)
plt.tight_layout()
```

# What is the Income / Age Relationship?

```python
sns.catplot(data=df, x='income', y='age', kind= 'box', aspect=1.5, palette="mako")
plt.xlabel('Income', size = 15)
plt.ylabel('Age', size = 15)
plt.title('Income / Age Relationship', size = 18)
plt.tight_layout()
```

Most of the people having income less than or equal to 50K are between age 25-45 and most of the people having income over 50K are between age 37-50


# Distribution Of Occupation?

```python
plt.figure(figsize=(20,10))
total = len(df['occupation'])
ax = sns.countplot(x='occupation', data=df)
for h in ax.patches:
    height = h.get_height()
    ax.text(h.get_x() + h.get_width() / 2., height + 3,'{:1.2f}'.format((height/total)*100) + '%',
            ha="center",size=17)
plt.yticks(size = 17)
plt.xticks(size = 17)
plt.xlabel('Occupation', size = 30)
plt.xticks(rotation=45)
plt.ylabel('Number Of Prople', size = 30)
plt.title('Distribution Of Occupation', size = 30)
plt.tight_layout()
```

We can observe 14 unique categories present in the occupation attribute.    
Prof-specialty has the maximum count but Craft-repair and Exec-managerial has comparable number of observations.     
Armed-Forces has minimum samples in the occupation attribute.

# Witch race on average making more ?

```python
mf = df.groupby('race')['income_2'].mean().sort_values()
fig, ax = plt.subplots(figsize=(8, 8))

patches, texts, pcts = ax.pie(mf, normalize=True, 
                              labels = ['Amer-Indian-Eskimo', 'Black', 'Other', 'White', 'Asian-Pac-Islander'], 
        autopct = '%1.1f%%', colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#ffbb66'],
       textprops={'size': 'x-large'}, wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'})


for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())

plt.setp(pcts, color='black')

plt.tight_layout()
```

It's look like Asian-Pac-Islander and White has the highest income compare to others races
